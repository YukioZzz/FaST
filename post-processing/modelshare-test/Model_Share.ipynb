```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns
sns.set_style("white")
# define the used rectangles
used_rectangles = [
    (0, 0, 0, 400, 12),
    (1, 400, 0, 400, 24),
    (2, 0, 24, 600, 12),
    (3, 0, 12, 200, 12),
    (4, 600, 24, 400, 24),
    (5, 0, 36, 600, 12),
    (6, 0, 48, 400, 12)
]

# define the free rectangles
free_rectangles = [
    (0, 200, 12, 200, 12),
    (1, 800, 0, 200, 24),
    (2, 0, 60, 1000, 40),
    (3, 400, 48, 600, 52)
]

# create a figure and axis
def plotPackingBin(used_rectangles, free_rectangles):
    fig, ax = plt.subplots(figsize=(10, 5))

    # set the x and y limits of the axis
    ax.set_xlim(0, 1000)
    ax.set_ylim(0, 100)
    palette = sns.color_palette("mako_r", 4)
    # loop over the used rectangles
    for i, x, y, w, h in used_rectangles:
        # create a rectangle patch
        rect = Rectangle((x, y), w, h, linewidth=1, edgecolor='black', facecolor=palette[0])

        # add the rectangle patch to the axis
        ax.add_patch(rect)

        # add the index number to the center of the rectangle
        ax.text(x + w/2, y + h/2, str(i), horizontalalignment='center', verticalalignment='center', color='white', fontsize=14, weight='bold')

    # loop over the free rectangles
    palette = sns.color_palette("Set1", len(free_rectangles))
    for i, x, y, w, h in free_rectangles:
        # create a rectangle patch
        rect = Rectangle((x, y), w, h, linewidth=3, edgecolor=palette[i], facecolor='none')
        # add the rectangle patch to the axis
        ax.add_patch(rect)
        # add the index number to the center of the rectangle
        ax.text(x + w/2, y + h/2, str(i), horizontalalignment='center', verticalalignment='center', color=palette[i], fontsize=14, weight='bold')

    # show the plot
    # plt.show()
    
plotPackingBin(used_rectangles, free_rectangles)
plt.savefig("FaSTNodeScheduler0.svg", format="svg")
```
